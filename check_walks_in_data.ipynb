```python
import pandas as pd
import numpy as np

from tqdm import tqdm
```

```python
def read_walks(path, walk_file_name):
    w_tuples = set()
    for line in tqdm(open(path + walk_file_name, 'r')):
        w = line.strip().split(' ')
        w = np.array(w, dtype=object)
        for i in range(len(w)-1):
            h = int(w[i][1:])
            t = int(w[i+1][1:])
            e = (h, t)
            w_tuples.add(e)
    return w_tuples
```

```python
def read_files(path, merge_file_name, mcols=['h','t','r', 'h_id', 't_id', 'r_id']):
    df_merge = pd.read_csv(path + merge_file_name
                                  , dtype={mcols[0]: object, mcols[1]: object, mcols[2]:object, mcols[3]:int
                                           , mcols[4]:int, mcols[5]:int})
    print('df_merge: {}'.format(df_merge.shape))
    
    e_tuples = set()
    for i in df_merge.itertuples():
        h = i[4]
        t = i[5]
        e = (h,t)
        e2 = (t, h)
        e_tuples.add(e)
        e_tuples.add(e2)
    print('e_tuples: {}'.format(len(e_tuples)))
        
    return e_tuples, df_merge
```

```python
walk_tuples = read_walks('data/', 'walks-pro-x-pro-path-40-40-graphpattern2vec-biased-v4_ALL-vCH_br')
```

```python
def compare_sets(s1,s2):
    
    d1 = s1-s2
    if len(d1) == 0:
        print('two sets are the same.')
    print('not in s1: {}'.format(len(d1)))
    
    d2 = s2-s1
    print('not in s2: {}'.format(len(d2)))
    
    return d1, d2
```
